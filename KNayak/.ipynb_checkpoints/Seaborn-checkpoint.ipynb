# Distribution Plots

* distplot
* joinplot
* pairplot

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.shape
```

```python
df_corr=df.corr()
fig,ax=plt.subplots(figsize=(10,6))
ax=sns.heatmap(df_corr,lw=0.2,fmt='.2f',cmap='winter',annot=True)
```

# Univariate Analysis

```python
sns.jointplot(x='tip',y='total_bill', data=df, height=10, kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill', data=df, height=10, kind='reg')
```

# Pair Plot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='smoker')
```

# Dist Plot

Dist plot helps us to check the distribution of the columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False)
```

# Count Plot

```python
sns.countplot('sex', data=df);
```

```python
sns.countplot('smoker', data=df);
```

```python
sns.countplot('day', data=df);
```

```python
sns.countplot(y='sex', data=df);
```

```python
sns.barplot(x = 'total_bill',y = 'sex',data=df)
```

```python
sns.barplot(x = 'smoker',y = 'total_bill',data=df)
```

# Box Plot

```python
sns.boxplot('smoker','total_bill', data=df);
```

```python
sns.boxplot('sex','total_bill', data=df);
```

```python
sns.boxplot('day','total_bill', data=df, palette='rainbow');
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

```python
sns.barplot(x='smoker',y='total_bill',hue='sex',data=df)
```

```python
sns.barplot(x='smoker',y='tip',hue='sex',data=df)
```

# Violin Plot

```python
sns.violinplot(x='total_bill',y='day', data=df)
```

```python
fig,ax= plt.subplots(figsize=(8,6))
ax = sns.violinplot(x='total_bill',y='day', data=df, palette='rainbow')
```
