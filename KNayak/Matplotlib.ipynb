```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,20,0.25)
x
```

```python
y = x**2
y
```

```python
plt.scatter(x,y, c='r')
plt.xlabel('Value of X')
plt.ylabel('Value of y');
plt.title('Plot of X vs Y')
```

```python
plt.plot(x,y, 'k+-', lw=1)
plt.xlabel('Value of X')
plt.ylabel('Value of y');
plt.title('Plot of X vs Y')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b');
```

# Bar Plot

```python
x = [2,8,10]
y = [11,16,9]

x2 = [3,6,9]
y2 = [7,12,2]
plt.bar(x,y)
plt.bar(x2,y2)
plt.title('Bar Graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')
```

# Box Plot with Matplotlib

```python
data = [np.random.normal(0,std,100)for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True);
```

# Pie Chart

```python
# Data to plot
labels = 'Python', 'C++','Ruby', 'Java'
sizes = [215,130,245,201]
plt.pie(sizes,labels=labels, autopct='%1.1f%%',shadow=False)
plt.show()
```
